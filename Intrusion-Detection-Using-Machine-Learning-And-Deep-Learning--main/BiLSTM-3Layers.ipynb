```python
import pandas as pd
```

```python
x1 = pd.read_csv(r'D:/FYP dataset/ARP MitM/ARP_MitM_dataset.csv',header=None)
x1
```

```python
y1 = pd.read_csv(r'D:/FYP dataset/ARP MitM/ARP_MitM_labels.csv')
y1
```

```python
x2 = pd.read_csv(r'D:/FYP dataset/SSDP Flood/SSDP_Flood_dataset.csv',header=None)
x2
```

```python
y2 = pd.read_csv(r'D:/FYP dataset/SSDP Flood/SSDP_Flood_labels.csv')
y2
```

```python
x = pd.concat([x1,x2])
x
```

```python
y = pd.concat([y1,y2])
y
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import tensorflow as tf
from keras.layers import Dense, BatchNormalization, Dropout, LSTM, Bidirectional
from keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from keras.optimizers import Adam
from tensorflow.keras import regularizers
from sklearn.metrics import precision_score, recall_score, confusion_matrix, classification_report, accuracy_score, f1_score
from keras import callbacks
from tensorflow.keras.callbacks import EarlyStopping
```

```python
y.drop('Unnamed: 0',inplace = True, axis=1)
y
```

```python
print("Number of 0's and 1's in y dataset:")
print(y['x'].value_counts())
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y,train_size = 0.8, test_size = 0.2, random_state = 0,stratify = y)
```

```python
x_train , y_train = np.array(x_train), np.array(y_train)
x_train = np.reshape(x_train, (x_train.shape[0] , x_train.shape[1], 1) )
```

```python
x_test , y_test = np.array(x_test), np.array(y_test)
x_test = np.reshape(x_test, (x_test.shape[0] , x_test.shape[1], 1) )
```

```python
model = Sequential()
model.add(Bidirectional(LSTM(units=5, return_sequences=True, input_shape = (x_train.shape[1],1) ) ))
model.add(Dropout(0.2))
model.add(LSTM(units= 5 , return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units= 5 , return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units= 5))
model.add(Dropout(0.2))
model.add(Dense(units = 1,activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['acc'])
history = model.fit(x_train, y_train, validation_split=0.1, epochs=5,batch_size=1028 )
```

```python
print(history.history.keys())

plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['train', 'test'], loc = 'upper left')
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['train', 'test'], loc = 'upper left')
plt.show()
```

```python
trainloss=model.evaluate(x_train,y_train,verbose=0)
testloss=model.evaluate(x_test,y_test,verbose=0)
```

```python
print("Train Loss: ",trainloss)
print("Test Loss: ",testloss)
```
